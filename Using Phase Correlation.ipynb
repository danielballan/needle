I suspect that the point-like features are anchoring the image. I would expect a *slightly* nonzero result even so, but maybe the rotating wire is completely domainted by these unmoving points. I will use the existing tools (developed for the "covariance" / "inertial axes" method) in `preprocessing.py` to select an ROI around the wire.

```python
from preprocessing import bigfish, pad_roi, threshold
```

```python
# This function:
# - Prepares the image (wire should be white, not black)
# - Finds the largest region and defining an "ideal" roi
# - Adjusts that roi to a standard-size "roi" since images must have equal size
# - Crops the *original* unmodified image

def crop_roi(image, size=40):
    # Get ideal ROI
    roi = bigfish(threshold(np.invert(image.astype('uint8'))))
    
    # Enlarge to standard ROI size, being as centered as possible but within image bounds.
    if np.any(size > np.array(image.shape)):
        raise ValueError("Standard ROI size must be less than every image dimension.")
    s0, s1 = roi  # slice objects
    size0 = s0.stop - s0.start
    start0 = (s0.start + size0 // 2 - size // 2).clip(0)
    stop0 = start0 + size
    size1 = s1.stop - s1.start
    start1 = (s1.start + size1 // 2 - size // 2).clip(0)
    stop1 = start1 + size
    
    # Crop
    return image[slice(start0, stop0), slice(start1, stop1)]
```

```python
# Test it out on two images.
fig, ax = plt.subplots()
ax.imshow(crop_roi(images[0]))

fig, ax = plt.subplots()
ax.imshow(crop_roi(images[5]))
```

```python
# Looks good: do them all.
cropped_images = map(crop_roi, images)
```
